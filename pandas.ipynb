##  Визуализация

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.pie([0.25, 0.30, 0.65], labels=['25%', '30%', '65%'], colors=['#34df56', 'b', (0.4, 0.0, 0.25)])
ax.set_title('Pie chart', fontsize=15)
```

```python
fig, ax = plt.subplots()

ax.bar(np.arange(0, 5, 1), np.random.randint(3, 10, 5))
ax.set_title('Bar plot', fontsize=15)
plt.xticks(np.arange(0, 5, 1), ['c','h','o','k','v']);
```

```python
fig, ax = plt.subplots()

ax.bar(np.arange(0, 5, 1), np.random.randint(3, 10, 5), width=0.25)
ax.bar(np.arange(0, 5, 1), np.random.randint(2, 4, 5), color='r', align='edge', width=0.4)
ax.set_title('Bar plot', fontsize=15)
plt.xticks(np.arange(0, 5, 1), ['c','h','o','k','v']);
```

```python
fig, ax = plt.subplots()

a = np.random.randint(3, 10, 5)
b = np.random.randint(2, 4, 5)
ax.bar(np.arange(0, 5, 1), a, width=0.25)
ax.bar(np.arange(0, 5, 1), b, bottom=a)

ax.set_title('Bar plot', fontsize=15)
plt.xticks(np.arange(0, 5, 1), ['c','h','o','k','v']);
```
